```python
def add_code(df, d_code, h_code, l_code):
    df = df.copy()   

    # D Code
    df['person_prefer_d_1_n'] = df['person_prefer_d_1'].apply(lambda x: d_code[x]['속성 D 세분류코드'])
    df['person_prefer_d_1_s'] = df['person_prefer_d_1'].apply(lambda x: d_code[x]['속성 D 소분류코드'])
    df['person_prefer_d_1_m'] = df['person_prefer_d_1'].apply(lambda x: d_code[x]['속성 D 중분류코드'])
    df['person_prefer_d_1_l'] = df['person_prefer_d_1'].apply(lambda x: d_code[x]['속성 D 대분류코드'])

    df['person_prefer_d_2_n'] = df['person_prefer_d_2'].apply(lambda x: d_code[x]['속성 D 세분류코드'])
    df['person_prefer_d_2_s'] = df['person_prefer_d_2'].apply(lambda x: d_code[x]['속성 D 소분류코드'])
    df['person_prefer_d_2_m'] = df['person_prefer_d_2'].apply(lambda x: d_code[x]['속성 D 중분류코드'])
    df['person_prefer_d_2_l'] = df['person_prefer_d_2'].apply(lambda x: d_code[x]['속성 D 대분류코드'])

    df['person_prefer_d_3_n'] = df['person_prefer_d_3'].apply(lambda x: d_code[x]['속성 D 세분류코드'])
    df['person_prefer_d_3_s'] = df['person_prefer_d_3'].apply(lambda x: d_code[x]['속성 D 소분류코드'])
    df['person_prefer_d_3_m'] = df['person_prefer_d_3'].apply(lambda x: d_code[x]['속성 D 중분류코드'])
    df['person_prefer_d_3_l'] = df['person_prefer_d_3'].apply(lambda x: d_code[x]['속성 D 대분류코드'])

    df['contents_attribute_d_n'] = df['contents_attribute_d'].apply(lambda x: d_code[x]['속성 D 세분류코드'])
    df['contents_attribute_d_s'] = df['contents_attribute_d'].apply(lambda x: d_code[x]['속성 D 소분류코드'])
    df['contents_attribute_d_m'] = df['contents_attribute_d'].apply(lambda x: d_code[x]['속성 D 중분류코드'])
    df['contents_attribute_d_l'] = df['contents_attribute_d'].apply(lambda x: d_code[x]['속성 D 대분류코드'])

    # H Code
    df['person_prefer_h_1_l'] = df['person_prefer_h_1'].apply(lambda x: h_code[x]['속성 H 대분류코드'])
    df['person_prefer_h_1_m'] = df['person_prefer_h_1'].apply(lambda x: h_code[x]['속성 H 중분류코드'])
    
    df['person_prefer_h_2_l'] = df['person_prefer_h_2'].apply(lambda x: h_code[x]['속성 H 대분류코드'])
    df['person_prefer_h_2_m'] = df['person_prefer_h_2'].apply(lambda x: h_code[x]['속성 H 중분류코드'])
    
    df['person_prefer_h_3_l'] = df['person_prefer_h_3'].apply(lambda x: h_code[x]['속성 H 대분류코드'])
    df['person_prefer_h_3_m'] = df['person_prefer_h_3'].apply(lambda x: h_code[x]['속성 H 중분류코드'])

    df['contents_attribute_h_l'] = df['contents_attribute_h'].apply(lambda x: h_code[x]['속성 H 대분류코드'])
    df['contents_attribute_h_m'] = df['contents_attribute_h'].apply(lambda x: h_code[x]['속성 H 중분류코드'])

    # L Code
    df['contents_attribute_l_n'] = df['contents_attribute_l'].apply(lambda x: l_code[x]['속성 L 세분류코드'])
    df['contents_attribute_l_s'] = df['contents_attribute_l'].apply(lambda x: l_code[x]['속성 L 소분류코드'])
    df['contents_attribute_l_m'] = df['contents_attribute_l'].apply(lambda x: l_code[x]['속성 L 중분류코드'])
    df['contents_attribute_l_l'] = df['contents_attribute_l'].apply(lambda x: l_code[x]['속성 L 대분류코드'])
    return df

train_data = add_code(train_data, d_code, h_code, l_code)
test_data = add_code(test_data, d_code, h_code, l_code)
print("train_data.shape: ", train_data.shape)
print("test_data.shape: ", test_data.shape)
```

```python
def preprocessing_contents_open_dt(data):
    data['contents_open_dt'] = data['contents_open_dt'].astype('str')
    DATE = data['contents_open_dt'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))
    
    DATE = pd.DataFrame(DATE)
    DATE = DATE.rename(columns = {'contents_open_dt': 'date'})
    
    DATE['Y'] = DATE['date'].apply(lambda x: x.timetuple()[0])
    DATE['M'] = DATE['date'].apply(lambda x: x.timetuple()[1])
    DATE['D'] = DATE['date'].apply(lambda x: x.timetuple()[2])
    DATE['id'] = data['id']
    
    data = data.merge(DATE, on = 'id', how = 'left')
    data = data.drop(columns = ['date', 'contents_open_dt'])
    return data

train_data = preprocessing_contents_open_dt(train_data)
test_data = preprocessing_contents_open_dt(test_data)

# 안전하게 확인하고 넘어 갑시다. 
train_data_labels = train_data['target']
train_data, test_data = train_data.align(test_data, join = 'inner', axis = 1)
train_data['target'] = train_data_labels
print("train_data.shape: ", train_data.shape)
print("test_data.shape: ", test_data.shape)
```

```python
header = list(train_data.head())
#len(header)
#print(header)

dic = {}

def make_dic(l, param):
  if param == 'itself':
    for i in l:
      dic[i] = i
    return
  for i in l:
    dic[i] = param

itself = ['id', 'person_rn', 'contents_rn', 'target']
date = ['Y', 'M', 'D']
ordinal = ['person_attribute_a_1', 'person_attribute_b', 'person_prefer_e', 'contents_attribute_e']
match = header[header.index('person_prefer_d_1_n') : header.index('contents_attribute_l_l')+1]
boolen = header[1:7]

make_dic(itself, 'itself')
make_dic(date, 'date')
make_dic(ordinal, 'ordinal')
make_dic(match, 'match')
make_dic(boolen, 'boolen')


for i in header:
  if i not in dic:
    dic[i] = 'norminal'
    
metadf = pd.DataFrame.from_dict(dic, orient = 'index')
print(metadf)
print(len(header))
metadic = dic
#df.to_csv("metadata.csv")
```

#Smooth

```python
def smooth(DF, output,dic,column, alpha):
  #DF : input df
  #dic : match_dict
  #column : smoothing 을 적용할 column
  #alpha : weight
  #return 값
  #match_dict : categorial값 - mean 값
  df = DF.copy()
  df = df[[column, "target"]]

  global_mean = df['target'].mean()
  Agg = df.groupby(column)['target'].agg(["count", "mean"])
  Smooth = (Agg["count"] * Agg["mean"] + alpha*global_mean)/(Agg["count"]+alpha)
  output.loc[:, column] = DF[column].map(Smooth)
  dic[column] = Smooth
```

```python
DF = job_data.copy()
match_dict = {} #catagory 값 - mean값


for i in categorial:
  smooth(DF, categorial_col_train, match_dict,i, 100) #alpha 값 0.7로

for i in categorial:
  categorial_col_test.loc[:, i] = job_data_test[i].map(match_dict[i])
```

```python
X_train = pd.concat([categorial_col_train,not_categorial_col_train],axis=1)
X_test=pd.concat([categorial_col_test,not_categorial_col_test],axis=1)
```

```python
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(X_train, target, test_size=.3)
```

```python
not_categorial_col_train = pd.DataFrame()
not_categorial_col_test = pd.DataFrame()
target = pd.DataFrame()
```

```python
DF = job_data.copy()
header = list(DF.columns)
categorial = []
not_categorial = []

for i in header:
  if metadic[i] == 'norminal' or metadic[i] == 'match':
    categorial.append(i)
  else:
    not_categorial.append(i)

not_categorial.remove('target')
```

```python
not_categorial_col_train =  not_categorial_col_train.append(job_data[not_categorial])
target = job_data['target']
not_categorial_col_test =  not_categorial_col_test.append(job_data_test[not_categorial])
```

```python
categorial_col_train = job_data[categorial].copy()
categorial_col_test = job_data_test[categorial].copy()
```

```python
from sklearn.model_selection import StratifiedKFold

def Cross_validation(df, output,column, fold):
  DF = df.copy()
  global_mean = DF["target"].mean()
  output[column+"_valid"] = np.nan

  X_train = DF.drop("target", axis = 1)
  Y_train = DF["target"]
  skf = StratifiedKFold(n_splits = fold, shuffle = True, random_state = 33)

  for tr_idx, val_idx in skf.split(X_train, Y_train):
    X_train, X_val = DF.iloc[tr_idx].copy(), DF.iloc[val_idx].copy()
    X_val[column+"_valid"] = X_val[column].map(X_train.groupby(column)["target"].mean())
    output[column+"_valid"].iloc[val_idx] = X_val[column+"_valid"]

  output[column+"_valid"] = output[column+"_valid"].fillna(global_mean)
```

```python
DF = job_data.copy()
match_dict = {}

for i in categorial:
  Cross_validation(DF, categorial_col_train, i, 10) #10 fold
  match_dict[i] = categorial_col_train.groupby(i)[i+"_valid"].mean()
```

```python
for i in categorial:
  categorial_col_test.loc[:, i] = job_data_test[i].map(match_dict[i])
```

```python
X_train = pd.concat([categorial_col_train,not_categorial_col_train],axis=1)
X_test=pd.concat([categorial_col_test,not_categorial_col_test],axis=1)

from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(X_train, target, test_size=.3)

from sklearn.ensemble import RandomForestClassifier
rf= RandomForestClassifier()
rf.fit(x_train, y_train)
rf.score(x_test,y_test)
```
